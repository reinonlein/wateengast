**Elk feestje begint met wat imports:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

**Met een beetje data: de 700 titels van alle posts tot nu toe**

```python
posts = pd.read_excel('700posts.xlsx', index=False)
posts.head()
```

**Vervolgens gooi je alles in één lange tekststring:**

```python
woordenbende = " ".join(str(titel) for titel in posts['Title'])
```

```python
print(f"Dat zijn dus in totaal {len(woordenbende)} woorden, verspreid over {len(posts)} posts.")
```

```python
wordcloud = (
    WordCloud(
    width=800, 
    height=400, 
    stopwords=['de', 'het', 'met', 'een', 'te', 'op', 'je', 'van', 'is', 'in'], 
    background_color="white", 
    colormap="prism")
    .generate(woordenbende)
)
```
